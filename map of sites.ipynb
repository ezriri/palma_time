```python
# this is plot of 1 section 
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# bounds for la-palma-nth:
# min lat: 28.8735
# max lat: 28.6785
# min long: -18.0176
# max long: -17.7209

df = pd.read_csv('transect_1_processed.csv',usecols=['Lat','Long']) ## just got lat + long

# boundaries of the map image:
BBox = ((df.Long.min(), df.Long.max(), df.Lat.min(),df.Lat.max()))
#print(BBox)

map = plt.imread('t1-specif-dim.png')

fig, ax = plt.subplots()
ax.scatter(df.Long, df.Lat,zorder = 1)
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(map, zorder = 0, extent = BBox)
```

```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# bounds for la-palma-nth:
# min lat: 28.8735
# max lat: 28.6785
# min long: -18.0176
# max long: -17.7209

df_1 = pd.read_csv('transect_1_processed.csv',usecols=['Lat','Long']) ## just got lat + long
df_2 = pd.read_csv('transect_2_processed.csv',usecols=['Lat','Long']) ## just got lat + long

# boundaries of the map image:
#BBox = ((df.Long.min(), df.Long.max(), df.Lat.min(),df.Lat.max()))
BBox = [-17.9295, -17.7722, 28.7104, 28.7967]

map = plt.imread('nth-palma.png')

fig, ax = plt.subplots()
ax.scatter(df_1.Long, df_1.Lat,zorder = 1, c = 'r')
ax.scatter(df_2.Long, df_2.Lat,zorder = 1, c = 'b')

ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(map, zorder = 0, extent = BBox)

# i think this is right, pic of the different sites
```
